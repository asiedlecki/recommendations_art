```python
import pandas as pd

df = pd.read_json(path_or_buf='datasets/similar_movies.json')
```

```python
movies_0 = 0
movies_1 = 0
movies_else = 0
movies = {}

for col in df.columns:
    movies.setdefault(col, {})
    num_0 = 0
    num_1 = 0
    num_else = 0
    for row in range(df.shape[0]):
        if df.loc[row][col][0] == 0.0:
            num_0 += 1
        elif df.loc[row][col][0] == 1.0:
            num_1 += 1
        else:
            num_else += 1
#             print('{0}: {1}'.format(col, df.loc[row][col]))
        
    movies[col]['0'] = num_0
    movies[col]['1'] = num_1
    movies[col]['else'] = num_else

for movie in movies:
    if movies[movie]['0'] > 0:
        movies_0 += 1
    elif movies[movie]['1'] > 0:
        movies_1 += 1
    elif movies[movie]['else'] > 0:
        movies_0 += 1

print('1.0 movies: {0}'.format(round(movies_0/len(df.columns)*100, 2)))
print('0.0: {0}'.format(round(movies_1/len(df.columns)*100, 2)))
print('else: {0}'.format(round(movies_else/len(df.columns)*100, 2)))
```
