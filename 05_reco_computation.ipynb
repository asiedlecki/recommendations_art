```python
interval = 5000 # every <interval> value data will be dumped to file and RAM cleared
for ptype in part_types:

    print('-----', datetime.datetime.now())
    set_file = '{train_test}_set_dict_u{min_user_ratings}_m{min_movie_ratings}_str_{stratify_by}_{levels}_ts{test_size}_ran{random_state}'\
            .format(train_test=ptype,
                    min_user_ratings=min_user_ratings,
                    min_movie_ratings=min_movie_ratings,
                    stratify_by=stratify_by,
                    levels='_'.join([str(x) for x in levels]),
                    test_size=int(test_size*100),
                    random_state=random_state)
    
    train_set = dataset.openJson('datasets/train_test_sets/{body}_train_is{inner_split_perc}_ran{random_state}.json'\
                                 .format(body=set_file,
                                         inner_split_perc=int(is_perc*100),
                                         random_state=random_state_2))
    
    i=0
    print(ptype, datetime.datetime.now())
    results_dict = {}
    for user in train_set:
        length = len(train_set)
        i+=1
        if i<=20000: 
            continue
        else:
            result = rec.getRecommendedItems_v3(user_prefs=train_set[user], movies_sim=trained_data)
            results_dict[user] = result
            if i%50 == 0:
                print(ptype, i, datetime.datetime.now())

            if i%interval == 0 or i == length:


                target_file = 'datasets/reco_sets/{train_test}_reco_u{min_user_ratings}_m{min_movie_ratings}_str_{stratify_by}_{levels}_ts{test_size}_ran{random_state}_train_is{inner_split_perc}_ran{random_state_2}_pt{part}'\
                        .format(train_test=ptype,
                                min_user_ratings=min_user_ratings,
                                min_movie_ratings=min_movie_ratings,
                                stratify_by=stratify_by,
                                levels='_'.join([str(x) for x in levels]),
                                test_size=int(test_size*100),
                                random_state=random_state,
                                inner_split_perc=int(is_perc*100),
                                random_state_2=random_state_2,
                                part=i)

                dataset.savePrefsToJson(target_file=target_file, prefs=results_dict)
                print(i, 'saved', datetime.datetime.now())   
                del results_dict
                results_dict = {}
    
    del train_set
    del results_dict
```
