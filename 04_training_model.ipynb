```python
import recommendations as rec
import similarity as sim
import dataset
```

```python
# on what data train a model?
# train set to load - parameters
min_user_ratings = 6
min_movie_ratings = 69
stratify_by = 'count'
levels = [16, 35, 102]
test_size = 0.2 # test size of initial split
random_state = 42 # random state of initial split
```

```python
# model parameters
most_popular=False
n=1000000
distance= (sim.pearson_numpy, 'pear')
use_buffer=True
```

```python
data_file = 'prefs_filtered_u{min_user_ratings}_m{min_movie_ratings}'\
    .format(min_user_ratings=min_user_ratings, min_movie_ratings=min_movie_ratings)

data = dataset.openJson(file='datasets/{prefs}.json'.format(prefs=data_file))
```

```python
data = dataset.transformPrefs(data)
```

```python
train_set_file = 'train_set_dict_u{min_user_ratings}_m{min_movie_ratings}_str_{stratify_by}_{levels}_ts{test_size}_ran{random_state}'\
                .format(min_user_ratings=min_user_ratings,
                        min_movie_ratings=min_movie_ratings,
                        stratify_by=stratify_by,
                        levels='_'.join([str(x) for x in levels]),
                        test_size=int(test_size*100),
                        random_state=random_state)
```

```python
train_set = dataset.openJson(file='datasets/train_test_sets/{train_set_file}.json'.format(train_set_file=train_set_file))
```

```python
prefs = {key:values for key, values in data.items() if key in train_set}
```

```python
len(prefs)
```

```python
len(train_set)
```

```python
prefs = dataset.transformPrefs(prefs)
```

```python
target_file = '''datasets/trained_data/similar_movies_u{min_user_ratings}_m{min_movie_ratings}_str_{stratify_by}_{levels}_ts{test_size}_ran{random_state}_n{n}_{distance}.json'''\
                             .format(min_user_ratings=min_user_ratings,
                                     min_movie_ratings=min_movie_ratings,
                                     stratify_by=stratify_by,
                                     levels='_'.join([str(x) for x in levels]),
                                     test_size=int(test_size*100),
                                     random_state=random_state,
                                     n=n,
                                     distance=distance[1])
```

```python
target_file
```

```python
rec.executeDictSimMovies(prefs_dict=prefs,
                         target_file=target_file,
                         most_popular=most_popular,
                         n=n,
                         distance=distance[0],
                         use_buffer=use_buffer)
```
